```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
```

```python
fig = plt.figure(); ax = fig.add_subplot(1,1,1)

ax.plot(np.random.randn(1000).cumsum(), 'k', label = 'one')

ax.plot(np.random.randn(1000).cumsum(), 'k--', label ='two')

ax.plot(np.random.randn(1000).cumsum(), 'k.', label = 'three')

ax.legend(loc = 'best')

plt.show()
```

# ways of interacting with the matplotlib configuration system
# the first is programmatically from python using the rc method
plt.rc('figure', figsize = (10,10))

# the first argument to rc is the component you wish to customize

font_options = {'family':'monospace','weight':'bold','size':'small'}
plt.rc('font', **font_options)

```python
# plotting functions in pandas
s = pd.Series(np.random.randn(10).cumsum(), index = np.arange(0,100,10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10,4).cumsum(0), columns = ['A','B','C','D'], index = np.arange(0,100,10))
df.plot()
```

```python
# making bar plots
# is a simple as passing kind = 'bar' (for vertical bars) or kind = 'barh' (for horizontal bars)
fig, axes = plt.subplots(2,1)

data = pd.Series(np.random.randn(16), index = list('abcdefghijklmnop'))

data.plot(kind = 'bar', ax = axes[0], color = 'k', alpha = 0.7)

data.plot(kind = 'barh', ax = axes[1], color = 'k', alpha = 0.7)
```
